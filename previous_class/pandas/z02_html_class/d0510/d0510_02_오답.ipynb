```python
def add_score(kor):
    kor = kor+20
    if kor > 100: kor=100
    return kor
```

```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
df.sort_index(ascending=True,inplace=True)
df
```

```python
# 총합,평균이라는 컬럼을 추가하시오
df['총합'] = 0
df['평균'] = 0
df['총합'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회'] 
df['평균'] = df['총합'] /5
```

```python
def grade(score):
    # score = df['평균']
    if score >=90: return 'A',
    elif score >=80: return 'B',
    elif score >=70: return 'C',
    elif score >=60: return 'D',
    else: return 'F'
```

#### lambda함수: 익명함수 -> lambda 매개변수
#### def(매개변수)

```python
def result(score):
    txt = ''
    if score>=70 : txt = '합격'
    else : txt= '불합격'
    return txt
```

```python
df['결과'] = df['평균'].apply(lambda score:'합격' if score >= 70 else '불합격')
#df[결과] = df[평균].apply(result)
```

```python
df
```

```python
# 평점이라는 컬럼을 만든 뒤 평균 90이면 A, 80 B...60미만 F 처리
df['평점']=df['평점'].apply(grade)
df
```

```python
# 결과2 컬럼을 생성 65점 초과 합격 65점 재시험 65점 미만 불합격
df['결과2']=df['평균'].apply(lambda score:'합격' if score>65 else ('재시험' if score==65 else '불합격'))
```

```python
df
```

```python
# 국어+영어 합쳐서 100점을 넘지 않으면 국어에 20점 추가

filt = df['국어']+df['영어'] 

df.loc[filt< 100,'국어']=df.loc[filt< 100,'국어']+20
df
```
