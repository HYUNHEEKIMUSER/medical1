```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
df.sort_index(ascending=True,inplace=True)
df
```

```python
# 사회 컬럼을 삭제 음악컬럼 추가
# 음악 점수는 과학점수가 80점이상이면 80, 80점 미만이면 65점 입력
# 합계 컬럼을 추가해서 국어 영어 수학 과학 음악으로 추가
# 컬럼의 위치는 음악 과학 바로 뒤로 컬럼 위치 변경
df.drop(labels='사회',axis=1)
```

```python
# 음악 점수는 과학점수가 80점이상이면 80, 80점 미만이면 65점 입력
# 합계 컬럼을 추가해서 국어 영어 수학 과학 음악으로 추가
df['음악'] = 65
df.loc[df['과학'] >=80,'음악'] = 80
a_list = list(df.columns)
# 컬럼의 위치는 음악 과학 바로 뒤로 컬럼 위치 변경
df = df[a_list[1:7]+[a_list[-1]]+a_list[7:-1]]
# 사회삭제
df.drop(labels='사회',axis=1,inplace=True)
df
```
