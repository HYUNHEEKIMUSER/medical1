
#### dropna - Nan 처리: 지우기


```python
df.dropna(inplace=True)
df
```

```python
# axis: index,column
# how: any, all
```

```python
# nan 들어가 있는 컬럼에서 모든 컬럼이 nan인 것만 삭제
df.dropna(axis='columns',how='all',inplace=True)
df
```

```python
# Nan 들어가 있는 컬럼에서 1개라도 nan이 포함되어 있으면 삭제
df.dropna(axis='columns',how='any',inplace=True)
df
```

```python
df.loc['4번','영어']=np.nan

```

```python
df.dropna(axis='columns',how='any',inplace=True)
```

```python
df
```

```python
# nan이 들어가 있는 index에서 한개라도 nan이 포함되어 있으면 삭제
df.dropna(axis='index',how='any',inplace=True)
df
```
