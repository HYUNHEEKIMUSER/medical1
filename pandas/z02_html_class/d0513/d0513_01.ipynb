```python
# 함수사용
# 3학년 중에 성적이 가장 낮은 학생 출력
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
df['합계'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df
df[df['학년']==3]
df[df['학년']==3]['합계'].min()
# df[(df['학년']==3) & (df['합계'].min())]
```

```python
# 3학년 중에 합계가 가장 낮은 점수를 찾아서, 그 조건으로 합계점수 조건을 가지고 row데이터 검색

df[df['합계']==df[df['학년']==3]['합계'].min()]
```

```python
# 평균 컬럼추가, grade 컬럼에 a..f학점을 출력하시오
df['평균'] = df['합계']/5
df['grade'] = 0
df
```

```python
def change(score):
    if score >=90: return 'A'
    elif score >=80: return 'B'
    elif score >=70: return 'C'
    elif score >=60: return 'D'
    else: return 'F'
    
```

```python
df['grade'] = df['평균'].apply(change)
df
```

```python
df['SW특기']=df['SW특기'].str.upper()
df['SW특기'].fillna('python',inplace=True)
df
```

```python
def upper_change(txt):
    if pd.isnull(txt):
        return "PYTHON"
    else:
        return txt.upper()
```

```python
df['SW특기'] = df['SW특기'].apply(upper_change)
df
```
