```python
# 학년별로 크기 조절 1*500=500, 2*500=1000, 3*500=1500
sizes = df['학년']*500
plt.scatter(df['영어'],df['수학'],s=sizes)
plt.xlabel('영어점수')
plt.ylabel('수학점수')
```

```python
sizes = df['학년']*500
plt.scatter(df['영어'],df['수학'],s=sizes,c=df['학년'],cmap='viridis')
plt.xlabel('영어점수')
plt.ylabel('수학점수')
# colorbar에 눈금표시: ticks, 라벨표시: label, 컬러바크기조절:shrink
plt.colorbar(ticks=[1,2,3],label = '학년',shrink=0.5)
```
