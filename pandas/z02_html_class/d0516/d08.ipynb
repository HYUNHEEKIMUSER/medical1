```python
import matplotlib.pyplot as plt
# 한글설정
import matplotlib 
import pandas as pd
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #윈도우사용자
# matplotlib.rcParams['font.family'] = 'AppleGothic Gothic' 맥사용자
matplotlib.rcParams['font.size'] = '15'  #글자크기
matplotlib.rcParams['axes.unicode_minus'] = False #마이너스 표시 해결
```

```python
# df = df.T 가로 세로 바꾸는 법
df = df.T
df
```

```python
# 출생아 수는 막대그래프로 합계출산율은 꺾은 선 그래프로 출력\
plt.figure(figsize=(13,5))
plt.bar(df.columns,df.loc['출생아 수'],color ='orange', label= '출생아 수(천명)')
a = df.loc['출생아 수']
for i, txt in enumerate(a):
    plt.text(df.columns[i],a[i]+5,txt,ha='center',color='black')
plt.ylabel('출생아 수 (천명)',labelpad=15)
plt.ylim(250,700)
plt.yticks([300,400,500,600])
plt.show()
```

```python
plt.figure(figsize=(13,5))
plt.plot(df.columns,df.loc['합계 출산율'],'y',ls='-',linewidth=5,marker='o',markeredgecolor='yellow',markerfacecolor='yellow',markersize = 10)
b = df.loc['합계 출산율']
for i, txt in enumerate(b):
    plt.text(df.columns[i],b[i]+0.05,txt,ha='center',color='black')
plt.ylabel('합계 출산율 (가임여성 1명당 명)',labelpad=15)
plt.ylim(0,1.5)
plt.yticks([0,1])
```

```python
plt.figure(figsize=(13,5))
plt.bar(df.index,df['출생아 수'])

# 값표시
for i,txt in enumerate(df['출생아 수']):
    plt.text(df.index[i],df['출생아 수'][i]+5,txt,ha='center')

plt.ylim(250,700)
plt.yticks([300,400,500,600])
plt.show()

# ylim : 0, 1.5
# yticks : 0,1
plt.figure(figsize=(13,5))
plt.plot(df.index,df['합계 출산율'],marker='o')
# 값표시
for i,txt in enumerate(df['합계 출산율']):
    plt.text(df.index[i],df['합계 출산율'][i]+0.05,txt,ha='center')
plt.ylim(0,1.5)
plt.yticks([0,1])

plt.show()



plt.plot(df.index, df['출생아 수'])
plt.plot(df.index, df['합계 출산율'])

fig, ax1 = plt.subplots(figsize=(10,5))

# 1번째 막대 그래프
ax1.set_ylabel('출생아 수')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
for i, txt in enumerate(df['출생아 수']):
    ax1.text(df.index[i],df['출생아 수'][i]+5,txt,ha='center')
ax1.bar(df.index,df['출생아 수'], color='#ff812d')

# 2번째 꺽은선 그래프
# 1개의 그래프에 내용을 2개 넣기
ax2 = ax1.twinx()
ax2.set_ylabel('합계 출산율')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index, df['합계 출산율'],color='#ffd100',marker='o',markersize=10,markeredgecolor='w',markeredgewidth=3,linewidth=3)

# 데이터 출력
for i,txt in enumerate(df['합계 출산율']):
    ax2.text(df.index[i],df['합계 출산율'][i]+0.05,txt,ha='center')
```
