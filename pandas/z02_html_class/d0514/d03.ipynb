```python
# chipotle.tsv에서 주문 중 가장 많이 주문한 메뉴 10개를 꺾은 선 그래프로 출력
import pandas as pd
df = pd.read_csv('chipotle.tsv',sep='\t')
df
```

```python
# item_name 별 주문한 갯수의 합은
# 주문별 그룹->quantity의 합을 구해야 함
x=df.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head(5)
```

```python
x1=df['item_name'].value_counts()[:5]
# y= df.head(10).groupby('item_name')['quantity'].max()
```

```python
plt.plot(x,label ='메뉴')
plt.plot(x1,label ='메뉴')
# plt.plot(x,y,label ='메뉴')
plt.title('많이 주문한 메뉴 10개')
plt.legend(loc='upper center')
```

```python
a=df.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head().index
b=df.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head().values
c=df['item_name'].value_counts()[:5].values
```

```python
plt.plot(a,b,label = '주문수')
plt.plot(a,c,label = '주문횟수')
plt.legend(loc='upper right')
```
