```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
df.head(3)
```

```python
# 키를 꺾은 선 그래프로 출력하시오
# 이름 키
# y축 170,180,190,200,210
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #윈도우사용자
# matplotlib.rcParams['font.family'] = 'AppleGothic Gothic' 맥사용자
matplotlib.rcParams['font.size'] = '15'  #글자크기
matplotlib.rcParams['axes.unicode_minus'] = False #마이너스 표시 해결
```

```python
x = df['이름']
y = df['키']
plt.title('키')
plt.plot(x,y,label = '학생키',marker = 'o',markeredgecolor='violet')
for i, txt in enumerate(y):
    plt.text(x[i],y[i]+1.7,txt,color='black',ha='center')
plt.legend(loc='upper right')
# plt.yticks([170,180,190,200,210])  #그래프에 있는 Y축 눈금표시
plt.ylim(160,215)
plt.show()
```
