```python
# SW특기 컬럼에서 JAVA가 포함되어 있는 열을 출력하세요
import pandas as pd
df = pd.read_excel('score.xlsx',index_col = '지원번호')
df.sort_index(ascending=True,inplace=True)   #지원번호(index) 가지고 정렬

df['SW특기'].str.lower().str.contains('java')
df.fillna("미선택",inplace=True)
df['SW특기']=df['SW특기'].str.upper()
df['학교']=df['학교']+'등학교'
df
```

```python
# 과학점수를 +5점을 더해서 출력하시오

df['과학']=df['과학'].astype(int)+5
df

df['국어'] = df['국어']+10
df
```
