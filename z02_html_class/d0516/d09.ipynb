```python
# 각 항목별 상관관계 함수 1과 가까운 수 일수록 상관관계가 있다고 판단
df.corr()
```

```python
data_corr = df.corr()
fig = plt.figure(figsize=(16,16))
sns.heatmap(data_corr,annot=True) #각 항목별 데이터를 출력시켜줌
plt.show()
```

```python
plt.title('outcome을 가지고 원그래프 그리기')
values = [df['Outcome'].mean(),df['Outcome'].max(),df['Outcome'].min()]
labels = ['평균','최대','최소']
plt.pie(values,labels=labels,autopct='%.1f%%',startangle=90,counterclock=False)
plt.show()
```

```python
df[df['Outcome']==1].count() #268
df['Outcome'].value_counts()  # 0은 500명 2은 268명
explore = [0,0.1]
values = df['Outcome'].value_counts().values
labels = ['비환자','환자']
plt.pie(values,labels=labels,autopct='%.1f%%',startangle=90,explode=explore)
plt.title('당뇨병 환자')
```
