```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
df.sort_index(ascending=True,inplace=True)
df
```

```python
df['총합'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회'] 
df['평균'] = df['총합'] /5
```

```python
def grade(score):
    if score >= 90 : return "A"
    elif score >= 80 : return "B"
    elif score >= 70 : return "C"
    elif score >= 60 : return "D"
    else: return "F"
```

```python
df['평점'] = df['평균'].apply(grade)
```

```python
df
```

```python
def result(score):
    txt = ''
    if score >= 70 : txt = "합격"
    else: txt = "불합격"
    return txt
```

```python
df['결과'] = df['평균'].apply(lambda score:'합격' if score >= 70 else '불합격')
```

```python
df
```

```python
# 국어 + 영어 = 합쳐서 100점을 넘지 않으면, 국어 +20점을 추가해서 출력하시오.
filt = df['국어'] + df['영어']
df.loc[filt<100,'국어'] = df.loc[filt<100,'국어']+20
```

```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
df.sort_index(ascending=True,inplace=True)
df
```

```python
df.fillna('Java',inplace=True)
```

```python
df['SW특기'].str.lower()
```

```python
def lower_change(lang):
    # lang null 아닌 지 묻는 것 / notnull(null이아닌지묻는것)  isnull(null인지묻는것)
    if pd.notnull(lang): return lang.lower()
    else: return 'python'
```

```python
# 1. sw특기에서 각각 단어를 소문자로 변경하고 nan에 있는 데이터는 python으로 입력해서 출력
df['SW특기'] = df['SW특기'].apply(lower_change)
df
```
