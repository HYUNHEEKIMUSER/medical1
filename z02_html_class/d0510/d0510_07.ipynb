```python
# 주문당 갯수가 가장 많은 상위 5개 출력하시오
import pandas as pd
df = pd.read_csv('chipotle.tsv',sep='\t')
df['item_price']=df['item_price'].str.replace("$","").astype(float)
df.groupby('order_id').sum().sort_values('quantity',ascending=False)[['quantity','item_price']][:5]
```
