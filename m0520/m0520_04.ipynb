2. 원핫 코드로 변환
(데이터의 가중치 때문에 정답률이 다를 수도 있음)

```python
target=[]
data = []
attr_list = []
for row_index, row in df.iterrows():
    target.append(row.loc[0]) #target 결과 데이터 저장
    row_data= []
    for col, v in enumerate(row.loc[1:]):
        if row_index == 0:
            attr = {'dic':{},"cnt":0} # cnt 종류수
            attr_list.append(attr)
        else:
            attr = attr_list[col] 
        # 원핫인코딩 변환
        d = [0]*12 
        if v in attr['dic']:
            idx = attr['dic'][v] 
            # print(idx)
        else :
            idx = attr['cnt']
            attr['dic'][v] = idx
            # print(attr['dic'][v])
            attr['cnt']+=1 
        d[idx]=1
        row_data +=d     #원핫인코딩의 데이터 저장파일
    data.append(row_data) # 1줄 리스트를 다시 리스트에 추가
data[0]
# df2 = pd.DataFrame(data)
# df2.head()
```

2-(1) 원핫 코드로 변환 후 정답률

```python
from sklearn.ensemble import RandomForestClassifier
# 데이터 전처리
train_input, test_input, train_target, test_target = train_test_split(data,target)
# 데이터 학습 시키기 (svm,SVC()와 같다)
clf=KNeighborsClassifier()
clf.fit(train_input,train_target)
# 데이터 예측하기
predict = clf.predict(test_input)
print('결과값: ',predict)
# 정답률 구하기
score = clf.score(test_input,test_target)
print('정답률: ',score)
```
