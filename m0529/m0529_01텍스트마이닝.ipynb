```python
okt = Okt()
n_list = okt.nouns(new1)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# 직접 말뭉치를 count하는 방법

counter = {}
for n in n_list: #말뭉치 list 
    if n not in counter: # counter dict '국회' 단어 있는 지 확인
        if len(n) <= 1: continue  # 길이가 한 글자인 것은 제외
        counter[n] = 0 # {'국회': 0, '마지막':0}
    counter[n] +=1 # {'국회': 1,  '마지막':1}
```

```python
# counter에 있는 횟수로 정렬
# operator.itemgetter(1): 두 번째 데이터로 정렬기준, 나온 횟수
counter2 = sorted(counter.items(), key=operator.itemgetter(1), reverse=True)
```

```python
# 클라우드 그리기 
# wordcloud: 한글 지원이 안됨
font_path = 'NanumBarunGothic.ttf'

wc = WordCloud(font_path=font_path,background_color='white', width=800, height= 600)
# counter2가 리스트인데 dict로 바꿈
cloud = wc.generate_from_frequencies(dict(counter2))
plt.figure(figsize=(10,8))
plt.axis('off') # 눈금 없애기
plt.imshow(cloud)
```
